# Exploratory Data Analysis (EDA)
Analysis of Financial Transactions Dataset for Anomaly Detection.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_theme(style="whitegrid")
```

## 1. Load Cleaned Data

```python
# We will load a sample of the cleaned transactions data to save memory during EDA
df_tx = pd.read_csv('../data/processed/transactions_data_cleaned.csv', nrows=1000000)
df_tx['date'] = pd.to_datetime(df_tx['date'])
df_tx.head()
```

## 2. Univariate Analysis
### Transaction Amounts

```python
plt.figure(figsize=(10, 5))
sns.histplot(df_tx[df_tx['amount'] < 500]['amount'], bins=50, kde=True)
plt.title('Distribution of Transaction Amounts (Under $500)')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()
```

### Transaction Types / Categories (MCC)

```python
plt.figure(figsize=(12, 6))
top_mcc = df_tx['mcc'].value_counts().head(20)
sns.barplot(x=top_mcc.index, y=top_mcc.values, order=top_mcc.index, palette='viridis')
plt.title('Top 20 Merchant Category Codes (MCC)')
plt.xlabel('MCC')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

## 3. Time-Based Patterns

```python
df_tx['hour'] = df_tx['date'].dt.hour
df_tx['day_of_week'] = df_tx['date'].dt.dayofweek

fig, axes = plt.subplots(1, 2, figsize=(15, 5))

sns.countplot(x='hour', data=df_tx, ax=axes[0], palette='Blues_d')
axes[0].set_title('Transactions by Hour of Day')

sns.countplot(x='day_of_week', data=df_tx, ax=axes[1], palette='Oranges_d')
axes[1].set_title('Transactions by Day of Week (0=Monday)')
axes[1].set_xticklabels(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])

plt.tight_layout()
plt.show()
```
